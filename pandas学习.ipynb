```python
import pandas as pd
import numpy as np
np.random.seed(42)
values = np.random.randint(50,101,10)
indexes = []
for i in range(1,11):
    indexes.append('学生'+str(i))
scores = pd.Series(values,indexes)
print(scores)

print(scores.mean())
print(scores.min())
print(scores.max())
print(len(scores[scores>scores.mean()]))
print(scores[scores>scores.mean()].count())
```

```python
#给定一周温度,找出超过30°C的天数,计算平均温,从高到低排序,找出温度变化最大的两天
import pandas as pd
import numpy as np
temp = pd.Series([28,31,29,32,30,27,33],index=['周一','周二','周三','周四','周五','周六','周日'])
print(temp[temp>30].count())
print(temp.mean())
print(temp.sort_values(ascending=False))

t = abs(temp.diff())
print(t.sort_values(ascending = False).keys()[:2].tolist())
print(*(t.sort_values(ascending = False).keys()[:2].tolist()))
```

```python
#给定10个交易日股票收盘价,计算每日收益率,找出最低最高收益率的日期,计算波动率(收益率的标准差)
prices = pd.Series([102.3,103.5,105.1,104.8,106.2,107.0,106.5,108.1,109.3,110.2],index=pd.date_range('2023-01-01',periods=10))
prices.pct_change()  #percent (今天/昨天)-1
a = prices.pct_change()
print(a)
print(a.max())
print(a.idxmax())
print(a.std())
```

```python
 #12个月销售量,计算季度平均销售量,最高销售量月份,月环比增长率,找出连续增长超过2个月的月份
sales = pd.Series([120,135,145,160,155,170,180,175,190,200,210,220],index=pd.date_range('2022-01-01',periods=12,freq='MS'))
print(sales)
sales.resample('QS').mean() #重新采样
print("最大销售额",sales.max())
print("最大销售额月份",sales.idxmax())
print("月环比增长率",sales.pct_change())
a = sales.pct_change()
b = a>0
print(b[b.rolling(3).sum()==3].keys().tolist())
```

```python
#学生成绩分析
#计算每位学生总分和平均分,找出数学高于90英语高于85,按从高到低排并输出前三

import pandas as pd
import numpy as np
scores = pd.DataFrame({'姓名':['张三','李四','王五','赵六','钱七'],
                   '数学':[85,92,78,88,95],'英语':[90,88,85,92,80],'物理':[75,80,88,85,90]
                   })
print(scores)
```

```python
 #每个学生的总分
scores[['数学','英语','物理']] #提取学科列
scores['总分']=scores[['数学','英语','物理']].sum(axis=1) #转化为横向轴
#scores['平均分']=scores['总分']/3
scores['平均分']=scores[['数学','英语','物理']].mean(axis=1)
scores
```

```python
scores[(scores['数学']>90)|(scores['英语']>85)]
```

```python
scores.sort_values('总分',ascending=False).head(3)
```

```python
scores.nlargest(3,columns=['总分'])
```
